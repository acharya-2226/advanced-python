## Question 8: Prime Number Function
Write `is_prime(n)` and display prime numbers between 1 and 50.


````python
# Function to find factors of n
def find_factors(n):
    factors = []
    for i in range(1, n + 1):  # Loop through numbers from 1 to n
        if n % i == 0:  # If i divides n with no remainder
            factors.append(i)
    return factors

# Check for prime numbers between 1 and 150
for i in range(1, 150):
    if len(find_factors(i)) == 2:  # Prime numbers have exactly two factors: 1 and the number itself
        print(i)
````

Then write a function that combines them and returns:
```
(name, total, average, result)
```

Explain why modular design is beneficial.


````python
# Function to calculate total marks
def calculate_total_marks(marks):
    return sum(marks)

# Function to calculate average marks
def calculate_average(marks):
    return sum(marks) / len(marks) if marks else 0

# Function to classify the result based on average marks
def classify_result(average):
    if average >= 75:
        return 'Distinction'
    elif average >= 50:
        return 'Pass'
    else:
        return 'Fail'

# Function that combines the results and returns a tuple
def combine_results(name, marks):
    total = calculate_total_marks(marks)
    average = calculate_average(marks)
    result = classify_result(average)
    
    return (name, total, average, result)

# Example usage
name = input
marks = [85, 90, 78, 88, 92]  # Example marks

result = combine_results(name, marks)
print(result)
````

## Question 10: Defensive Input Parsing
Write a function that accepts a raw input string in the format:

```
"Ram|101|22|78, 88, 91"
```
Hint: Parse student data from string format: "Name|Roll|Age|Mark1, Mark2, Mark3"

Tasks:
1. Extract name, roll number, age, and marks  
2. Convert them into appropriate data types  
3. Validate age (≥16) and marks (0–100)  
4. Return the result as a dictionary using roll number as the key  

The function must not crash for malformed input.


````python
def parse_student(raw_input):
    result = {}
    try:
        # Split by '|'
        parts = raw_input.split('|')
        if len(parts) != 4:
            raise ValueError("Input does not have exactly 4 parts")
        
        name = parts[0].strip()
        roll = parts[1].strip()
        age = int(parts[2].strip())
        marks_str = parts[3].strip()
        
        # Validate age
        if age < 16:
            raise ValueError(f"Age {age} is less than 16")
        
        # Convert marks to list of integers
        marks = [int(m.strip()) for m in marks_str.split(',')]
        
        # Validate marks
        for m in marks:
            if not (0 <= m <= 100):
                raise ValueError(f"Mark {m} is out of range 0-100")
        
        # Return dictionary with roll as key
        result[roll] = {
            'Name': name,
            'Age': age,
            'Marks': marks
        }
        
    except Exception as e:
        result['error'] = str(e)
    
    return result

# Example usage
input_str = "Ram|101|22|78, 88, 91"
print(parse_student(input_str))
````

## Question 11: Type Stability and Conversion
Given the list:
```python
data = [10, "20", 30.5, "abc", None, True]
```

Write a function that:
1. Separates numeric and non-numeric values  
2. Converts numeric values to float  
3. Returns both lists  

Explain in comments why `True` behaves as a numeric value.


````python
def separate_and_convert(data):
    numeric_values = []
    non_numeric_values = []
    
    for item in data:
        # Check for numeric values: int and float (bool is subclass of int in Python)
        if isinstance(item, (int, float)) and not isinstance(item, bool):
            numeric_values.append(float(item))  # Convert to float
        else:
            # Attempt to convert string to float
            try:
                numeric_values.append(float(item))
            except (ValueError, TypeError):
                non_numeric_values.append(item)
    
    return numeric_values, non_numeric_values

# Example usage
data = [10, "20", 30.5, "abc", None, True]
nums, non_nums = separate_and_convert(data)
print("Numeric:", nums)
print("Non-numeric:", non_nums)
````

## Question 14: Dictionary Aggregation Logic
Given:
```python
records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}
```

Write a function that:
1. Computes average per student  
2. Identifies top and bottom performers  
3. Returns results using a single dictionary


````python
def analyze_records(records):
    result = {}
    averages = {}
    
    # 1. Compute average per student
    for student, marks in records.items():
        avg = sum(marks) / len(marks)
        averages[student] = avg
    
    # 2. Identify top and bottom performers
    top_performer = max(averages, key=averages.get)
    bottom_performer = min(averages, key=averages.get)
    
    # 3. Return everything in a single dictionary
    result['Averages'] = averages
    result['Top Performer'] = {top_performer: averages[top_performer]}
    result['Bottom Performer'] = {bottom_performer: averages[bottom_performer]}
    
    return result

# Example usage
records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}

print(analyze_records(records))

````

## Question 15: Nested Data Structure Analysis
Given a nested dictionary representing a school's grade book:
```python
school_data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}
```

Write a function that:
1. Calculates the average marks for each student
2. Finds the top performer in each class
3. Calculates the overall school average
4. Returns a dictionary with:
   - `class_toppers`: {class_name: (student_name, average)}
   - `school_average`: float
   - `student_averages`: {student_name: average}

Explain in comments why nested dictionaries are useful for hierarchical data.

````python
def analyze_school(school_data):
    student_averages = {}
    class_toppers = {}
    total_sum = 0
    total_count = 0
    
    for class_name, students in school_data.items():
        class_avg_dict = {}
        for student, marks in students.items():
            avg = sum(marks) / len(marks)
            student_averages[student] = avg
            class_avg_dict[student] = avg
            
            # Accumulate for overall school average
            total_sum += sum(marks)
            total_count += len(marks)
        
        # Top performer in the class
        top_student = max(class_avg_dict, key=class_avg_dict.get)
        class_toppers[class_name] = (top_student, class_avg_dict[top_student])
    
    # Overall school average
    school_average = total_sum / total_count if total_count > 0 else 0
    
    # Return results in a structured dictionary
    return {
        "class_toppers": class_toppers,
        "school_average": school_average,
        "student_averages": student_averages
    }

# Example usage
school_data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}

result = analyze_school(school_data)
print(result)
````
