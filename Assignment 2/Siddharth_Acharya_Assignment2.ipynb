## Question 2: Variable Scope and Global
Create a global variable `counter = 0`.
- Write a function `increment_counter()` that modifies this global variable by increasing it by 1 each time the function is called.
- Call the function 5 times and print the final value of `counter`.

````python
# Write your answer here
counter = 0
def increase_counter():
	global counter 
	counter += 1
	
def decrease_counter():
	global counter
	counter -= 1
	
print("Initially counter is: ", counter)
increase_counter()
increase_counter()
increase_counter()
decrease_counter()

print("Final counter value after 3 increments and 1 decrement is: ", counter)
````

## Question 4: List Comprehension (Basics)

Use list comprehension to create a new list containing only the squares of the even numbers.

````python
# Write your answer here
# List Comprehension

squares_list_comprehension = [x**2 for x in range(1, 11) if x % 2 == 0]  # Alternate could be a loop from 2 to range 11 with step 2 or initialize with 0, step 2 and remove the 0 later
print("The squares of even numbers from 1 to 10 using list comprehension are: ", squares_list_comprehension)

# For reference through loop method
squares_loop = []
for x in range(1, 11):
    if x % 2 == 0:
        squares_loop.append(x**2)
print("Using loop, the squares of even numbers from 1 to 10 are: ", squares_loop)
# Here both methods give the same result but one is more concise, 
# uses only one line and is more advanced in terms of python features.

# Corresponding list of n numbers without squaring
numbers = [x for x in range(1, 11) if x % 2 == 0]
print("The corresponding even numbers from 1 to 10 are: ", numbers)
````

## Question 6: Dictionary Comprehension
Given two lists:
```python

```
Use dictionary comprehension to create a dictionary mapping student names to their marks.

````python
# Write your answer here
# Dictionary Comprehension

students = ["Alice", "Bob", "Charlie"]
marks = [85, 92, 78]

# Dictionary  Comprehension
students_marks = {students[i]: marks[i] for i in range(len(students))}

print("Students and their corresponding maks: ", students_marks)

# Using Loop for reference

students_marks_loop = {}
for i in range(len(students)):
    students_marks_loop[students[i]] = marks[i]
    
print("Using loop, Students and their corresponding marks: ", students_marks_loop)
````

## Question 11: Basic Exception Handling
Write a function `safe_divide(a, b)` that divides `a` by `b`.
- Use a `try-except` block to handle `ZeroDivisionError`.
- If a division by zero occurs, return "Cannot divide by zero" instead of crashing.

````python
# Write your answer here

# Basic Exception Handling to handle division by zero : safe division function

def safe_divide(a, b):
	try:
		result = a / b
	except ZeroDivisionError:
		return "Error: Division by zero is not allowed."
	else:
		return result
num1 = float(input("Enter numerator: "))
num2 = float(input("Enter denominator: "))
print(safe_divide(num1, num2))
````

- Handle `ValueError` if the input is not a number.
- Handle `TypeError` if the input is not a string/integer.
- Ensure the program prints a friendly error message for both cases.

````python
# Write your answer here
# Handling multiple exceptions

def safe_operations(a, b):
	try:
		division = a / b
		addition = a + b
	except ZeroDivisionError:
		return "Error: Division by zero is not allowed."
	except TypeError:
		return "Error: Invalid input type. Please enter numbers."
	else:
		return division, addition
num1 = input("Enter first number: ")
num2 = input("Enter second number: ")
try:
	num1 = float(num1)
	num2 = float(num2)	
except ValueError:
	print("Error: Invalid input type. Please enter numbers.")
else:
	print(safe_operations(num1, num2))
````

## Question 13: The 'Finally' Block
Write a code block that opens a file named `test.txt` (you can create a dummy file) in read mode.
- Use `try` to read the content.
- Use `finally` to ensure the file is closed properly, regardless of whether an error occurred during reading.

````python
# Write your answer here

#  Finally block example
def read_file(file_path):
	try:
		file = open(file_path, 'r')
		content = file.read()
		print("File content:\n", content)
	except FileNotFoundError:
		print("Error: The file was not found.")
	finally:
		try:
			file.close()
			print("File has been closed.")
		except NameError:
			print("File was never opened, so nothing to close.")
file_path = input("Enter the file path to read: ")
read_file(file_path)
````

## Question 15: OOP - Inheritance
Create a parent class `Employee` with attributes `name` and `salary`.
- Create a child class `Manager` that inherits from `Employee` and adds an attribute `department`.
- Write a method in `Manager` to display the name, salary, and department.

````python
# Write your answer here

# child class manager inheriting from parent class Employee

class Employee:
	def __init__(self, name, emp_id, salary):
		self.name = name
		self.emp_id = emp_id
		self.salary = salary

class Manager(Employee):
	def __init__(self, name, emp_id, salary, department):
		super().__init__(name, emp_id, salary)
		self.department = department

# Example of creating an object of the Manager class
manager1 = Manager("Siddharth Acharya", 'HCE080BCT037', 75000, "IT")
print("Manager Name:", manager1.name)

print("Employee ID:", manager1.emp_id)
print("Salary:", manager1.salary)
print("Department:", manager1.department)
````

## Question 16: OOP - Encapsulation
Create a class `BankAccount` with a private attribute `_balance`.
- Implement methods `deposit(amount)` and `withdraw(amount)` to modify the balance.
- Ensure that `withdraw` does not allow taking out more money than the current balance.
- Create a `get_balance()` method to safely access the private balance.

````python
# Write your answer here
# encaptulation : bankaccount with private attribute money and public methods to deposit, withdraw and check balance and validate money withdrawn < balance

class BankAccount:
	def __init__(self):
		self.__money = 0  # Private attribute

	def deposit(self, amount):
		if amount > 0:
			self.__money += amount
			print(f"Deposited: {amount}. New balance: {self.__money}")
		else:
			print("Deposit amount must be positive.")

	def withdraw(self, amount):
		if amount > self.__money:
			print("Insufficient funds for withdrawal.")
		elif amount <= 0:
			print("Withdrawal amount must be positive.")
		else:
			self.__money -= amount
			print(f"Withdrew: {amount}. New balance: {self.__money}")

	def check_balance(self):
		return self.__money
	
# Example usage
account = BankAccount()
account.deposit(500)
account.withdraw(200)
print("Current balance:", account.check_balance())

# Validation Trial
account.withdraw(400) 
````

## Question 18: Static Methods
Create a class `MathUtils` with a static method `is_even(n)`.
- The method should return `True` if `n` is even, and `False` otherwise.
- Call this method without creating an instance of the class.

````python
# Write your answer here

# example : mathutils with a static method is_even

class MathUtils:
	@staticmethod
	def is_even(number):
		return number % 2 == 0
# Example usage
num = int(input("Enter a number to check if it is even: "))
if MathUtils.is_even(num):
	print(f"{num} is even.")
else:
	print(f"{num} is odd.")
````

## Question 19: Variable Arguments (*args and **kwargs)
Write a function `student_report` that accepts:
- A mandatory argument `name`.
- Arbitrary positional arguments (`*args`) for subject scores.
- Arbitrary keyword arguments (`**kwargs`) for additional details (like `city`, `age`).
- The function should print the name, calculate the average of the scores, and print the additional details.

````python
# Write your answer here

# # args and kwargs example function : student_report


def student_report(name, *args, **kwargs):
	print(f"Student Name: {name}")
	if args:
		average_score = sum(args) / len(args)
		print(f"Average Score: {average_score:.2f}")
	else:
		print("No scores provided.")
	
	if kwargs:
		print("Additional Details:")
		for key, value in kwargs.items():
			print(f"{key}: {value}")
	else:
		print("No additional details provided.")
# Example usage
student_report("Siddharth Acharya", 85, 90, 78, city="Kathmandu", age=20)
````

## Question 20: Comprehensive System
Design a simple "Library Management System" using OOP and Exception Handling.
- Create a class `Library` with a list of available books.
- Add methods to `borrow_book(book_name)` and `return_book(book_name)`.
- If a user tries to borrow a book that isn't in the list, raise a custom exception `BookNotFoundError` (or use a standard `ValueError`).
- Ensure the state of the library updates correctly after each transaction.

````python
# Write your answer here
# Demonstration of a simple library systwm using oop and exception handling

class Library:
	def __init__(self, books):
		self.available_books = books

	def borrow_book(self, book_name):
		if book_name in self.available_books:
			self.available_books.remove(book_name)
			print(f"You have borrowed '{book_name}'.")
		else:
			raise ValueError(f"Book '{book_name}' not found in the library.")

	def return_book(self, book_name):
		self.available_books.append(book_name)
		print(f"You have returned '{book_name}'.")
# Example usage
library = Library(["1984", "To Kill a Mockingbird", "The Great Gatsby"])
try:
	library.borrow_book("1984")
	library.borrow_book("The Catcher in the Rye")  # This book is not in the library
except ValueError as e:
	print(e)
finally:
	try:
		library.return_book("1984")
	except ValueError as e:
		print(e)
````
